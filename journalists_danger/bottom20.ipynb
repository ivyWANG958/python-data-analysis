```python
import pandas as pd
```

```python
df = pd.read_csv('bottom20.csv')
```

```python
df = df[(df.legend != 'Total Score') & (df.legend != 'Status') ]
df.head()
```

```python
# df_num = df.apply(pd.to_numeric)
df_num = df[['Sudan',
 'Rwanda',
 'Saudi Arabia',
 'Zimbabwe',
 'Vietnam',
 'Somalia',
 'Laos',
 'China',
 'Libya',
 'IOTPA',
 'Syria',
 'Iran',
 'Myanmar',
 'Equatorial Guinea',
 'Belarus',
 'Uzbekistan',
 'Eritrea',
 'Cuba',
 'Turkmenistan',
 'North Korea']].apply(pd.to_numeric)
```

```python
df = df[['Year','legend']].join(df_num)
```

```python
country = ['Sudan',
 'Rwanda',
 'Saudi Arabia',
 'Zimbabwe',
 'Vietnam',
 'Somalia',
 'Laos',
 'China',
 'Libya',
 'IOTPA',
 'Syria',
 'Iran',
 'Myanmar',
 'Equatorial Guinea',
 'Belarus',
 'Uzbekistan',
 'Eritrea',
 'Cuba',
 'Turkmenistan',
 'North Korea']
```

```python
df_A = df[(df.legend == 'A-Legal')]
df_B = df[(df.legend == 'B-Political')]
df_C = df[(df.legend == 'C-Economic')]
```

```python
df_A_score = df_A.drop(['Year','legend'], axis = 1).sum(axis = 0, skipna = True)
bottom_a = df_A_score.sum(axis = 0) / 20/ 30 # scale 0-30
bottom_a
```

```python
df_B_score = df_B.drop(['Year','legend'], axis = 1).sum(axis = 0, skipna = True)
bottom_b = df_B_score.sum(axis = 0) / 20/ 40 # scale 0-40
bottom_b
```

```python
df_C_score = df_C.drop(['Year','legend'], axis = 1).sum(axis = 0, skipna = True)
bottom_c = df_C_score.sum(axis = 0) / 20/ 30 # scale 0-30
bottom_c
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,12))
for i in country:
    plt.plot(df_A[i], label= i )
    plt.legend()

plt.title('A-Legal score by year_bottom')
plt.xlabel('year')
plt.ylabel('score')
```
