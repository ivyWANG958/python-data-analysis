```python
import pandas as pd
```

```python
df = pd.read_csv('top20.csv')
```

```python
df = df[(df.legend != 'Total Score') & (df.legend != 'Status') ]
df.head()
```

```python
df_copy = df.copy()
```

```python
# df_num = df.apply(pd.to_numeric)
df_num = df[['Norway',
 'Sweden',
 'Finland',
 'Belgium',
 'Denmark',
 'Iceland',
 'Switzerland',
 'Luxembourg',
 'Netherlands',
 'Andorra',
 'Liechtenstein',
 'Palau',
 'New Zealand',
 'Saint Lucia',
 'Monaco',
 'Marshall Islands',
 'Ireland',
 'San Marino',
 'Portugal',
 'Estonia']].apply(pd.to_numeric)

#df[['Venezuela','Chad']].apply(pd.to_numeric)
```

```python
df = df[['Year','legend']].join(df_num)
```

```python
country = ['Norway',
 'Sweden',
 'Finland',
 'Belgium',
 'Denmark',
 'Iceland',
 'Switzerland',
 'Luxembourg',
 'Netherlands',
 'Andorra',
 'Liechtenstein',
 'Palau',
 'New Zealand',
 'Saint Lucia',
 'Monaco',
 'Marshall Islands',
 'Ireland',
 'San Marino',
 'Portugal',
 'Estonia']
```

```python
df_A = df[(df.legend == 'A-Legal')]
df_B = df[(df.legend == 'B-Political')]
df_C = df[(df.legend == 'C-Economic')]
```

```python
df_A_score = df_A.drop(['Year','legend'], axis = 1).sum(axis = 0, skipna = True)
bottom_a = df_A_score.sum(axis = 0) / 20/ 30 # scale 0-30
bottom_a
```

```python
df_B_score = df_B.drop(['Year','legend'], axis = 1).sum(axis = 0, skipna = True)
bottom_b = df_B_score.sum(axis = 0) / 20/ 40 # scale 0-40
bottom_b
```

```python
df_C_score = df_C.drop(['Year','legend'], axis = 1).sum(axis = 0, skipna = True)
bottom_c = df_C_score.sum(axis = 0) / 20/ 30 # scale 0-30
bottom_c
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))

for i in country:
    plt.plot(df_A[i], label= i )
    plt.legend()

plt.title('A-Legal score by_year top')
plt.xlabel('year')
plt.ylabel('score')
```

```python
plt.figure(figsize=(12,8))
for i in country:
    plt.plot(df_B[i], label= i )
    plt.legend()

    #B-Political	C-Economic	
plt.title('B-Political score by_year top')
plt.xlabel('year')
plt.ylabel('score')
```

```python
plt.figure(figsize=(12,8))
for i in country:
    plt.plot(df_C[i], label= i )
    plt.legend()

    #B-Political	C-Economic	
plt.title('C-Economic score by_year top')
plt.xlabel('year')
plt.ylabel('score')
```
