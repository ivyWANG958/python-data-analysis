```python
import pandas as pd
```

```python
df = pd.read_csv('bottom20-35.csv')
```

```python
df = df[(df.legend != 'Total Score') & (df.legend != 'Status') ]
df.head()
```

```python
df_copy = df.copy()
```

```python
# df_num = df.apply(pd.to_numeric)
df_num = df[['United Arab Emirates','Venezuela','Chad','Burundi','Brunei','Ethiopia','Russia','Bahrain','The Gambia','Swaziland','Yemen','Tajikistan','Kazakhstan','Azerbaijan','Congo (Kinshasa)']].apply(pd.to_numeric)

#df[['Venezuela','Chad']].apply(pd.to_numeric)
```

```python
df = df[['Year','legend']].join(df_num)
```

```python
df['Chad'].sum()
```

```python
country = ['United Arab Emirates',
 'Venezuela',
 'Chad',
 'Burundi',
 'Brunei',
 'Ethiopia',
 'Russia',
 'Bahrain',
 'The Gambia',
 'Swaziland',
 'Yemen',
 'Tajikistan',
 'Kazakhstan',
 'Azerbaijan',
 'Congo (Kinshasa)']
```

```python
df_A = df[(df.legend == 'ALegal')]
df_B = df[(df.legend == 'BPolitical')]
df_C = df[(df.legend == 'CEconomic')]
```

```python
df_A_score = df_A.drop(['Year','legend'], axis = 1).sum(axis = 0, skipna = True)
bottom_a = df_A_score.sum(axis = 0) / 15/ 30 # scale 0-30
bottom_a
```

```python
df_B_score = df_B.drop(['Year','legend'], axis = 1).sum(axis = 0, skipna = True)
bottom_b = df_B_score.sum(axis = 0) / 15/ 40 # scale 0-40
bottom_b
```

```python
df_C_score = df_C.drop(['Year','legend'], axis = 1).sum(axis = 0, skipna = True)
bottom_c = df_C_score.sum(axis = 0) / 15/ 30 # scale 0-30
bottom_c
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))

for i in country:
    plt.plot(df_A[i], label= i )
    plt.legend()

plt.title('A-Legal score by year_bottom')
plt.xlabel('year')
plt.ylabel('score')
```

```python
plt.figure(figsize=(12,8))
for i in country:
    plt.plot(df_B[i], label= i )
    plt.legend()

    #B-Political	C-Economic	
plt.title('B-Political score by year_bottom')
plt.xlabel('year')
plt.ylabel('score')
```

```python
plt.figure(figsize=(12,8))
for i in country:
    plt.plot(df_C[i], label= i )
    plt.legend()

    #B-Political	C-Economic	
plt.title('C-Economic score by year_bottom')
plt.xlabel('year')
plt.ylabel('score')
```
