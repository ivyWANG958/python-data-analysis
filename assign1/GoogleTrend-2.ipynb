```python
from selenium import webdriver
import requests
import csv
from bs4 import BeautifulSoup
import time
```

```python
url = 'https://trends.google.com/trends/trendingsearches/daily?geo=US'
browser = webdriver.Chrome()
browser.get(url)
time.sleep(2)
```

```python
def get_articles_from_browser():
    h = browser.find_element_by_css_selector('html')
    t = h.get_attribute('innerHTML')  
    mypage = BeautifulSoup(t)
   
    articles = []
    for content in mypage:
        rows = content.findAll("div",{"class":"feed-item"})
        for row in rows:
            details = row.findAll('div',{'class':'details-wrapper'})
            for detail in details:  
                article = {}
                index_container = detail.find('div',{'class':"index"})
                article['rank'] = index_container.text.strip()
                title_container = detail.find('div',{'class':"title"})
                article['title'] = title_container.text.strip()
                summary_container = detail.find("div",{"class":"summary-text"})
                article['summary'] = summary_container.text.strip()
                search_container = detail.find("div",{"class":"search-count-title"})
                article['search'] = search_container.text.strip()
                source_container = detail.find('div',{'class':"source-and-time"})
                article['source'] = source_container.text.strip().replace("•","|")
                link_container = detail.find('div',{'class':"summary-text"})
                article['link'] = link_container.a['href']
                articles.append(article)
            
    #print(articles)
    return articles
```

```python
data = []

while True:
    try:
        loadMoreButton = browser.find_element_by_class_name("feed-load-more-button")
        time.sleep(2)
        loadMoreButton.click()
        time.sleep(5)
    except Exception as e:
        print(e)
        break

data.extend(get_articles_from_browser())
        
print("Complete")
# time.sleep(10)
browser.quit()
```

```python
with open('Topsearch.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    header= ["rank", "title","search","summary","source","link"]
    spamwriter.writerow(header)
    spamwriter.writerows(data)
```
